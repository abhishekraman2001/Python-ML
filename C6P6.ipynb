```python
stu_details = {'sid':[1,2,3,4,5],
              'sname':['A','B','C','D','E'],
              'region':['N','S','E','E','W']}

stu_marks = {'sid':[1,2,3,4,6],
            'sub':['Py','R','Py','Tab','Py'],
            'marks':[43,23,44,22,32]}
```

```python
df1 = pd.DataFrame(stu_details)
df2 = pd.DataFrame(stu_marks)
print(df1)
print(df2)
```

```python
pd.merge(df1, df2, on='sid', how='outer')
```

```python
stu_details = {'s_id':[1,2,3,4,5],
              'sname':['A','B','C','D','E'],
              'region':['N','S','E','E','W']}

stu_marks = {'sid':[1,2,3,4,6],
            'sub':['Py','R','Py','Tab','Py'],
            'marks':[43,23,44,22,32]}
df1 = pd.DataFrame(stu_details)
df2 = pd.DataFrame(stu_marks)
print(df1)
print(df2)
```

```python
pd.merge(df1, df2, left_on='s_id', right_on='sid', how='outer')
```
